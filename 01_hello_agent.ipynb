# Module 01 — Hello, Agent!

This is the simplest possible Strands agent. No tools, no memory, no hooks — just a model and a message.

### What is Strands?

**Strands Agents** is an open-source SDK from AWS that lets you build AI agents with minimal boilerplate. Under the hood it:
- Manages the conversation loop (user → model → tool → model → ...)
- Handles tool calling automatically
- Integrates with Amazon Bedrock models
- Provides hooks for lifecycle events

The core primitive is the `Agent` class. Let's use it.

```python
from strands import Agent

# Create the simplest possible agent
# By default it uses Claude 3.7 Sonnet via Amazon Bedrock
agent = Agent()

# Invoke it — just like calling a function
response = agent("Hello! Who are you and what can you do?")
print(response)
```

That's it. One import, one object, one call.

---

### Choosing a model

You can specify any Bedrock model using its **inference profile ID** (note the `us.` prefix):
- `us.anthropic.claude-haiku-4-5-20251001-v1:0` (Claude Haiku 4.5 - fast, affordable)
- `us.anthropic.claude-3-5-haiku-20241022-v1:0` (Claude 3.5 Haiku - legacy)
- For Sonnet/Opus models, use the same `us.` prefix pattern

```python
# Use Claude Haiku 4.5 — faster and cheaper, great for workshops
agent = Agent(
    model="us.anthropic.claude-haiku-4-5-20251001-v1:0"
)

response = agent("What is 2 + 2? Answer in one sentence.")
print(response)
```

### Multi-turn conversation

The agent keeps conversation history automatically. Each call to `agent(...)` continues the same thread.

```python
agent = Agent(model="us.anthropic.claude-haiku-4-5-20251001-v1:0")

agent("My name is Yeshwanth.")
agent("I live in Mysuru.")

# The agent remembers the conversation context within the same session
response = agent("What do you know about me so far?")
print(response)
```

### Inspecting the agent

You can look at the messages the agent has accumulated.

```python
# See the raw conversation history
for msg in agent.messages:
    role = msg['role']
    content = msg['content']
    # content is a list of content blocks
    text = content[0].get('text', '') if content else ''
    print(f"[{role.upper()}]: {text[:100]}..." if len(text) > 100 else f"[{role.upper()}]: {text}")
```

---

### Key takeaways

- `Agent()` is all you need to get started
- Pass `model=` to choose a Bedrock model
- Calling `agent("...")` continues the conversation
- `agent.messages` holds the full history
