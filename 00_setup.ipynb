### Step 2 — Enable Amazon Bedrock Model Access

1. Go to the [AWS Console → Amazon Bedrock](https://console.aws.amazon.com/bedrock)
2. Navigate to **Model access** in the left sidebar
3. Request access to **Claude 3.5 Haiku** (and optionally Claude 3.5 Sonnet)
4. Wait for approval (usually instant for most regions)

---

### Step 3 — Configure AWS Credentials

Run `aws configure` in your terminal, or set environment variables:

```bash
export AWS_ACCESS_KEY_ID=your_key
export AWS_SECRET_ACCESS_KEY=your_secret
export AWS_DEFAULT_REGION=us-east-1
```

Or use an IAM role if running on EC2/Cloud9.

### Step 5 — Verify your setup
